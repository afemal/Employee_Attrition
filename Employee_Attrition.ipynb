## Problem Statement

Employee attrition is the natural process in which employees leave the worceforce. This can include retirement, voluntary, or involuntary separations from a company. Whether an employee choose to leave a company on his/her own accord, high attrition rates can be harmful to a company. They can cause:

- Incurred costs to the company as a result of recruitment, hiring, and training of new employees
- Decreased product or service quality
- Decreased employee performance or morale
- Delays in current projects or failure to meet deadlines
 
It is in the company's best interest to retain their current employees. If they can determine which employees are at risk of leaving, they can implement different techniques or incentives to reduce the risk. These may include additional training, wage increases, opportunities for promotion, etc.

## Proposal
I will use the HR Analytics Case Study dataset to create two classifications models to understand if the dataset has the ability to accurately classify whether an employee will stay or leave the company. This supervised learning problem is a binary classification problem with two labels Yes and No, indicating whether an employee has left the company or not. I will use two different metrics to evaluate the performance of my model: Recall and Area Under the Precision-Recall Curve. 

Recall is an important metric to consider because I want to minimize Type 2 errors. These errors indicate false negative predictions, which causes employees who are at risk of attrition to go undetected. Area Under the Precision-Recall Curve allows me to focus on the trade-off of precision and recall, as my data has class imbalance. Both metrics have scores between 0 and 1, where higher is better.

## Feature Selection 

I will perform feature selection using two different techniques. First, I will be using RFECV to conduct recursive feature elimination using cross-validation on two different classification models, Random Forest Classifier and Decision Tree Classifier. I will be using recall as the scoring metric. Second, I will use ANOVA F-values to remove irrelevant features. 

Before I perform feature selection, I will convert the categorical features into dummy variables using one hot encoding and the target to binary, with No=0 and Yes=1.

```python
def categorical_features(feature):
    cat = feature[cat_vars]
    cat_dummies = pd.get_dummies(cat, drop_first=True)
    feature = pd.concat([feature[num_vars], cat_dummies], axis=1)
    return feature

def binary_target(var):
    if (var == 'No'):
        return 0
    else:
        return 1
```

```python
x_train = categorical_features(x_train)
y_train = y_train.map(binary_target)                              
```

**RFECV Method**

Random Forest Classifier with recall scoring metric

```python
rf = RandomForestClassifier()
rf_rfecv_recall = RFECV(estimator=rf, step=1, scoring='recall')
rf_rfecv_recall.fit(x_train, y_train)
rf_rfecv_recall.transform(x_train)
```

```python
cols = rf_rfecv_recall.get_support(indices=True)
feature_selection_1 = x_train.iloc[:,cols]
feature_selection_1.columns
```

Decision Tree Classifier with recall scoring metric

```python
dt = DecisionTreeClassifier()
dt_rfecv_recall = RFECV(estimator=dt, step=1, scoring='recall')
dt_rfecv_recall.fit(x_train, y_train)
dt_rfecv_recall.transform(x_train)
```

```python
cols = dt_rfecv_recall.get_support(indices=True)
feature_selection_2 = x_train.iloc[:,cols]
feature_selection_2.columns
```

**ANOVA F-value Method**

```python
# I want to choose the best k that provides the best recall and average precision
for i in range(2, x_train.shape[1]):
    feature_selector = SelectKBest(f_classif, k=i)
    features_kbest = feature_selector.fit_transform(x_train, y_train)
    randomforest = RandomForestClassifier(random_state=seed)
    decisiontree = DecisionTreeClassifier(random_state=seed)
    
    model_rf = randomforest.fit(features_kbest, y_train)
    pred_rf = model_rf.predict(features_kbest)
    recall_rf = recall_score(y_train, pred_rf)
  
    model_dt = decisiontree.fit(features_kbest, y_train)
    pred_dt = model_dt.predict(features_kbest)
    recall_dt = recall_score(y_train, pred_dt)
   
    plt.scatter(i, recall_rf, c='b')
    plt.scatter(i, recall_dt, c='r')
   
```

It appears after k=9, recall has minimal increases. 

```python
fvalue_selector = SelectKBest(f_classif, k=9)
features_kbest = fvalue_selector.fit_transform(x_train, y_train)
cols = fvalue_selector.get_support(indices=True)
feature_selection_3 = x_train.iloc[:,cols]
feature_selection_3.columns
```

## Model Selection

```python
rf_pipeline = Pipeline([('rf', RandomForestClassifier(random_state=seed, n_jobs=-1))])

params = {'rf__n_estimators': [10, 20, 50, 100, 300, 500], 
               'rf__max_features': ['sqrt', 'log2'],
               'rf__max_depth': [4, 6, 8, 10, 12], 
               'rf__class_weight': ['balanced', None]}


rf_cv = GridSearchCV(rf_pipeline, 
                     param_grid=params, 
                     scoring={'avr': make_scorer(average_precision_score, needs_proba=True),
                              'll': make_scorer(log_loss, greater_is_better=False, needs_proba=True),
                              'recall': make_scorer(recall_score),
                              'f1_score': make_scorer(f1_score),
                              'roc_auc': make_scorer(roc_auc_score)}, 
                     n_jobs=-1,
                     cv=5,
                     refit='recall')
```
